```python
import pandas as pd
import numpy as np

# 合并数据
df = df_assets_weekly.merge(df_gas, left_index=True, right_index=True)

# Step 1: 计算 log return（稳定性更好）
log_returns = np.log(df.drop(columns=["Gas_Price"]) / df.drop(columns=["Gas_Price"]).shift(1))

# Step 2: 计算每个资产和油价的全区间 Pearson 相关系数（不随时间变）
correlations = {}
for col in df.columns.drop("Gas_Price"):
    correlations[col] = df["Gas_Price"].corr(df[col])

correlation_df = pd.DataFrame(correlations, index=["Correlation_with_Gas"]).T

# Step 3: Loop over each row (starting from second row due to shift) to calculate weight
rolling_weights = []

alpha1 = 0.7
alpha3 = 0.3

for i in range(1, len(log_returns)):
    current_date = df.index[i]
    current_return = log_returns.iloc[i]

    # 计算评分
    score = alpha1 * current_return + alpha3 * correlation_df["Correlation_with_Gas"]

    # 标准化为权重
    weight = score / score.sum()

    rolling_weights.append({
        "Date": current_date,
        **weight.to_dict()
    })

# 转成 DataFrame
rolling_weights_df = pd.DataFrame(rolling_weights).set_index("Date")
```

```python
# 初始化存储结果的字典
normalized_top_weights = {}

for date, row in rolling_weights_df.iterrows():
    # 只选取正权重资产
    positive_weights = row[row > 0]
    # 找出前 5 个正权重资产
    top_assets = positive_weights.sort_values(ascending=False).head(5)
    # 归一化，使这5个权重加总为1
    normalized_weights = top_assets / top_assets.sum()
    # 存储结果
    normalized_top_weights[date] = normalized_weights

# 转为 DataFrame
normalized_top_weights_df = pd.DataFrame.from_dict(normalized_top_weights, orient="index")

# 展示结果
print(normalized_top_weights_df)
```

```python
# 1. 初始化风险结果存储
portfolio_risk = {}

# 2. 遍历每一个时间点
for date in normalized_top_weights_df.index:
    # 当前这期的资产组合和对应的权重
    weights = normalized_top_weights_df.loc[date].dropna()
    assets = weights.index

    # 取出这些资产在当前时间点前一段时间的 log returns（建议使用过去 3~4 周）
    # 这里只取过去4期数据用于估算协方差矩阵
    available_dates = log_returns.index[log_returns.index < date]
    recent_dates = available_dates[-4:]  # 你可以改成 3 或 5

    # 获取对应资产的 log returns 子集
    recent_returns = log_returns.loc[recent_dates, assets]

    # 计算协方差矩阵
    cov_matrix = recent_returns.cov()

    # 计算组合风险 = sqrt(w.T @ Cov @ w)
    portfolio_std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

    # 存储
    portfolio_risk[date] = portfolio_std

# 3. 转为 DataFrame 查看
portfolio_risk_df = pd.DataFrame.from_dict(portfolio_risk, orient="index", columns=["Portfolio_StdDev"])
print(portfolio_risk_df)
```
