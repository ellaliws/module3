```python
%%writefile app.py
import streamlit as st
import numpy as np

# ---------------- Configuration ----------------
allowed_directions = ["up", "down"]
allowed_magnitudes = [5, 10, 15, 20, 25, 30, 40, 50, 75, 100]
max_amount = 10000  # in USD

# Parameters
MAGNITUDE_REF = 100
AMOUNT_REF = 10000
alpha = 0.6           # weight for magnitude
beta = 0.4            # weight for amount
trust_scale = 15      # controls decay of trust as deviation increases

# ---------------- Scoring Function ----------------
def compute_P_user(magnitude, amount, alpha=alpha, beta=beta):
    mag_score = np.log1p(min(magnitude, MAGNITUDE_REF)) / np.log1p(MAGNITUDE_REF)
    amt_score = np.log1p(min(amount, AMOUNT_REF)) / np.log1p(AMOUNT_REF)
    P_user = alpha * mag_score + beta * amt_score
    return round(min(max(P_user, 0), 1), 4)

# ---------------- Trust Function ----------------
def compute_trust_factor(user_magnitude, system_magnitude, scale=trust_scale):
    delta = abs(user_magnitude - system_magnitude)
    trust = np.exp(-delta / scale)
    return round(trust, 4)

# ---------------- System Forecast ----------------
# For demonstration, we set these values statically.
# In practice, these may come from your predictive model.
system_direction = "up"       # Example forecast direction
system_magnitude = 20         # Example forecast magnitude (in percent)

# ---------------- Streamlit Interface ----------------
st.title("Oil Price Trend Prediction")

# User selects forecast direction
direction = st.selectbox("Please predict the oil price trend in the next 3 months:", allowed_directions)

# User chooses the predicted percentage change
magnitude = st.selectbox("Enter your predicted price change (%)", allowed_magnitudes)

# User enters the betting amount
amount = st.number_input(f"Enter the amount you want to bet (USD, max {max_amount}):",
                         min_value=0.0, max_value=float(max_amount), value=0.0, step=1.0)

# When the user clicks "Submit", process the input
if st.button("Submit"):
    is_aligned = (system_direction == direction)

    # Compute risk preference and trust factor
    P_user = compute_P_user(magnitude, amount)
    trust_factor = compute_trust_factor(magnitude, system_magnitude)

    # ---------------- Output Summary ----------------
    st.markdown("## Prediction Summary")
    st.markdown(f"**System Forecast :** {system_direction} {system_magnitude}%")
    st.markdown(f"**Your Direction  :** {direction}")
    st.markdown(f"**Your Magnitude  :** {magnitude}%")
    st.markdown(f"**Bet Amount      :** ${amount}")
    st.markdown(f"**Risk Preference (P_user) :** {P_user} (0 = very conservative, 1 = highly risk-seeking)")
    st.markdown(f"**Trust in Your Prediction :** {trust_factor} (0 = no trust, 1 = fully trust)")
```

```python
import numpy as np

# ------------------ λ from P (exponential decay) ------------------ #
def lambda_from_P(P_user):
    """
    Calculate risk aversion coefficient λ based on user risk preference score P ∈ [0, 1].
    Lower P → higher λ; Higher P → lower λ.
    """
    λ = round(5 * np.exp(-2 * P_user), 2)
    return λ


# ------------------ Risk Classification ------------------ #
def classify_risk_level(P_user):
    """
    Classify user's risk level and assign appropriate optimization model and λ.
    Returns a dictionary with:
    - label: internal tag
    - name: readable name
    - model: optimization method
    - lambda: risk aversion parameter (None for Max Sharpe)
    """
    if P_user <= 0.33:
        return {
            "label": "averse",
            "name": "Risk Averse",
            "model": "CVaR",
            "lambda": lambda_from_P(P_user)
        }
    elif P_user < 0.66:
        return {
            "label": "neutral",
            "name": "Risk Neutral",
            "model": "Mean-Variance",
            "lambda": lambda_from_P(P_user)
        }
    else:
        return {
            "label": "seeking",
            "name": "Risk Seeking",
            "model": "Max Sharpe",
            "lambda": None  # not applicable
        }


# Classification
risk_info = classify_risk_level(P_user)

# Display Result
print(f"User Confidence P        : {P_user}")
print(f"Risk Type                : {risk_info['name']}")
print(f"Recommended Model        : {risk_info['model']}")
if risk_info["lambda"] is not None:
    print(f"Risk Aversion λ          : {risk_info['lambda']}")
else:
    print("Risk Aversion λ          : N/A (not used in this model)")
```

```python
def allocate_assets_v2(asset_df, P_user, trust_factor, risk_info, is_aligned, top_k=None):
    """
    Enhanced asset allocation:
    - P_user determines risk preference;
    - trust_factor (0~1) controls how much weight is given to user input;
    - final ratio is blended between user-suggested and system-default.
    """
    # Step 0: Determine top_k
    if top_k is None:
        top_k = {"averse": 10, "neutral": 8, "seeking": 5}.get(risk_info["label"], 8)

    # Step 1: Compute user-suggested positive ratio
    label = risk_info["label"]
    if label == 'averse':
        user_ratio = 1 - 0.5 * P_user if is_aligned else 1 - P_user
    elif label == 'neutral':
        user_ratio = 0.7 if is_aligned else 0.5
    elif label == 'seeking':
        user_ratio = 1.0 if is_aligned else 1 - P_user
    else:
        raise ValueError("Invalid risk label")

    # Step 2: Define system baseline ratio
    system_ratio = 0.8 if is_aligned else 0.5

    # Step 3: Blend user + system using trust factor
    pos_ratio = trust_factor * user_ratio + (1 - trust_factor) * system_ratio

    num_pos = round(top_k * pos_ratio)
    num_neg = top_k - num_pos

    # Step 4: Select top-N assets from positive/negative pools
    pos_pool = asset_df[asset_df["direction"] == "positive"].sort_values(by="normalized_score", ascending=False)
    neg_pool = asset_df[asset_df["direction"] == "negative"].sort_values(by="normalized_score", ascending=False)

    pos_assets = pos_pool.head(num_pos)
    neg_assets = neg_pool.head(num_neg)

    return pos_assets.reset_index(drop=True), neg_assets.reset_index(drop=True)


# 调用函数
pos_assets, neg_assets = allocate_assets_v2(asset_df, P_user, trust_factor,risk_info, is_aligned)

# 打印结果确认
print("Top K:", len(pos_assets) + len(neg_assets))
print("\nPositive Assets:")
print(pos_assets)

print("\nNegative Assets:")
print(neg_assets)
```

```python
import numpy as np


# === Risk Seeking: Max Sharpe Ratio ===
def utility_max_sharpe(expected_returns, cov_matrix, weights):
    """
    Max Sharpe Ratio:
    U(w) = E[r(w)] / Std(w)
    """
    expected = np.dot(weights, expected_returns)
    std_dev = np.sqrt(np.dot(weights, np.dot(cov_matrix, weights)))
    if std_dev == 0:
        return np.inf
    return -expected / std_dev  # Negative for minimization


# === Risk Neutral: Mean-Variance Utility ===
def utility_mean_variance(expected_returns, cov_matrix, weights, lambda_):
    """
    Mean-Variance Utility:
    U(w) = E[r(w)] - λ * Var(w)
    """
    expected = np.dot(weights, expected_returns)
    variance = np.dot(weights, np.dot(cov_matrix, weights))
    return -(expected - lambda_ * variance)


# === Risk Averse: CVaR Utility ===
def utility_cvar(expected_returns, losses_matrix, weights, lambda_, alpha=0.95):
    """
    CVaR Utility:
    U(w) = E[r(w)] - λ * CVaR_α(w)
    """
    portfolio_losses = np.dot(losses_matrix, weights)
    var_alpha = np.percentile(portfolio_losses, 100 * alpha)
    cvar_alpha = portfolio_losses[portfolio_losses >= var_alpha].mean()
    expected = np.dot(weights, expected_returns)
    return -(expected - lambda_ * cvar_alpha)


# === Objective Function Dispatcher ===
def get_objective_function(risk_info, expected_returns=None, cov_matrix=None, losses_matrix=None, score_vector=None):
    """
    Given risk_info and required inputs, return the corresponding utility function.
    Returns: callable objective_function(w)
    """
    risk_label = risk_info["label"]
    lambda_ = risk_info["lambda"]

    if risk_label == 'seeking':
        print("Risk Strategy: Max Sharpe Ratio")
        print("Objective: maximize E[r(w)] / Std(w)")
        return lambda w: utility_max_sharpe(expected_returns, cov_matrix, w)

    elif risk_label == 'neutral':
        print("Risk Strategy: Mean-Variance")
        print(f"Objective: maximize E[r(w)] - λ * Var(w)  (λ = {lambda_})")
        return lambda w: utility_mean_variance(expected_returns, cov_matrix, w, lambda_)

    elif risk_label == 'averse':
        print("Risk Strategy: CVaR Optimization")
        print(f"Objective: maximize E[r(w)] - λ * CVaR_α(w)  (λ = {lambda_})")
        return lambda w: utility_cvar(expected_returns, losses_matrix, w, lambda_, alpha=0.95)

    else:
        raise ValueError("Invalid risk label: expected 'seeking', 'neutral', or 'averse'")

```

```python

from scipy.optimize import minimize

# === Step 0: Input ===
# portfolio_assets: DataFrame with ['asset_name', 'expected_return', 'score']
# asset_price: historical price DataFrame, index as dates, columns as asset names
# risk_info: dict with 'label' and 'lambda'

expected_returns = portfolio_assets['expected_return'].values
score_vector = portfolio_assets['score'].values
asset_names = portfolio_assets['asset_name'].values

# Historical returns
price_df = asset_price[asset_names].dropna()
returns_matrix = price_df.pct_change().dropna()
cov_matrix = returns_matrix.cov().values
losses_matrix = -returns_matrix.values

n_assets = len(portfolio_assets)
lambda_ = risk_info["lambda"]
risk_label = risk_info["label"]

# === Step 1: Define Utility Functions ===

def utility_sharpe(w):
    port_return = np.dot(w, expected_returns)
    port_vol = np.sqrt(np.dot(w, np.dot(cov_matrix, w)) + 1e-6)
    return -port_return / port_vol

def utility_mean_variance(w):
    port_return = np.dot(w, expected_returns)
    port_var = np.dot(w, np.dot(cov_matrix, w))
    return -(port_return - lambda_ * port_var)

def utility_cvar(w, alpha=0.95):
    portfolio_losses = losses_matrix @ w
    var_threshold = np.percentile(portfolio_losses, 100 * alpha)
    cvar = portfolio_losses[portfolio_losses >= var_threshold].mean()
    port_return = np.dot(w, expected_returns)
    return -(port_return - lambda_ * cvar)


# === Step 2: Constraints ===
x0 = np.ones(n_assets) / n_assets
bounds = [(0, 1)] * n_assets
constraints = [{'type': 'eq', 'fun': lambda w: np.sum(w) - 1}]

# === Step 3: Primary Strategy Optimization ===
if risk_label == 'seeking':
    objective_func = utility_sharpe
elif risk_label == 'neutral':
    objective_func = utility_mean_variance
elif risk_label == 'averse':
    objective_func = utility_cvar
else:
    raise ValueError("Invalid risk label")

result_main = minimize(objective_func, x0=x0, bounds=bounds, constraints=constraints)
weights_main = result_main.x
return_main = np.dot(weights_main, expected_returns)


# === Step 4: Output DataFrame with Risk Metrics ===
port_variance = np.dot(weights_main, np.dot(cov_matrix, weights_main))
port_std = np.sqrt(port_variance)

result_df = pd.DataFrame({
    "Asset": asset_names,
    "Weight": weights_main,
    "Expected_Return": expected_returns,
})

# 添加总组合信息行
portfolio_summary = pd.DataFrame({
    "Asset": ["[Portfolio]"],
    "Weight": [np.sum(weights_main)],
    "Expected_Return": [return_main],
})

# 合并资产 + 组合总结表格（可选）
full_result_df = pd.concat([result_df, portfolio_summary], ignore_index=True)

print(full_result_df)

# === Step 5: Output Summary ===
print("\n=== Portfolio Summary ===")
print(f"Primary Strategy Expected Return : {return_main:.4f}")
print(f"Portfolio Volatility (Std. Dev.) : {port_std:.4f}")

```
